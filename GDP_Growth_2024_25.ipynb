**Quarterly GDP growth data for India (FY 2024-25)**

```python
import matplotlib.pyplot as plt

# Quarterly GDP growth data for India (FY 2024-25), approximated for illustration
quarters = ['Q1 2024-25', 'Q2 2024-25', 'Q3 2024-25', 'Q4 2024-25']
gdp_growth = [7.8, 6.2, 5.5, 7.4]  # Hypothetical data based on the narrative

# Plotting the timeline chart
plt.figure(figsize=(10, 6))
plt.plot(quarters, gdp_growth, marker='o', linestyle='-', color='mediumblue')
plt.title('India’s Quarterly GDP Growth in FY 2024–25')
plt.xlabel('Quarter')
plt.ylabel('GDP Growth (%)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(0, 9)
plt.axhline(y=6.5, color='green', linestyle='--', label='Annual Average (6.5%)')
plt.legend()
plt.tight_layout()

plt.show()
```
